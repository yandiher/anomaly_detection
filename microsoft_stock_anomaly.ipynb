# import necessary library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from adtk.detector import *
from adtk.visualization import plot
```

# download stock

```python
df = yf.download('MSFT', start='1995-01-01')['Adj Close']
```

```python
df.info()
```

# set detector

```python
detector = InterQuartileRangeAD(c=1)
anomalies = detector.fit_detect(df)
```

# plot the analysis

```python
plot(ts=df, 
     anomaly=anomalies,
     anomaly_color='red',
     anomaly_tag='marker',
     ts_linewidth=0.01)
plt.show()
```
