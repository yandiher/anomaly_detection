# import necessary library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from adtk.visualization import plot
from adtk.detector import *
```

# import data

```python
df = pd.read_csv('data/monthly_csv.csv')
df['Date'] = pd.to_datetime(df['Date'], yearfirst=True)
df = df.set_index('Date')
df = df['Mean']
```

```python
df.plot(figsize=(16,4))
plt.show()
```

# set detector

### quantile

```python
detector = QuantileAD(low=0.01, high=0.99)
anomalies = detector.fit_detect(df)
```

```python
plot(df, anomaly=anomalies,  anomaly_color='red', anomaly_tag='marker')
plt.show()
```

### interquartile

```python
detector = InterQuartileRangeAD(c=1)
anomalies = detector.fit_detect(df)
```

```python
plot(df, anomaly=anomalies,  anomaly_color='red', anomaly_tag='marker')
plt.show()
```

### persist AD

```python
detector = PersistAD(c=2)
anomalies = detector.fit_detect(ts=df)
```

```python
plot(df, anomaly=anomalies,  anomaly_color='red', anomaly_tag='marker')
plt.show()
```
